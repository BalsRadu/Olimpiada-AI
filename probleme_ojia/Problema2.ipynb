```python
import numpy as np
import pandas as pd
```

```python
df_train = pd.read_csv("02_data/train_data.csv")
df_test = pd.read_csv("02_data/test_data.csv")
```

```python
df_train.info()
df_train.describe(include="all")
```

```python
df_train["Timestamp"] = pd.to_datetime(df_train["Timestamp"])
df_train["hour"] = df_train["Timestamp"].dt.hour
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

feature_cols = [
    "Suspicious_Port_Activity",
    "Traffic_Volume_Variation",
    "Packet_Length_Anomaly",
    "Malware_Score",
    "Threat_Level_Index",
    "User_Behavior_Score",
    "Geo_Dispersion",
    "Payload_Entropy",
    "Session_Duration",
    "Packet_Retry_Rate",
    "Anomaly_Tendency",
    "Login_Attempts",
    "Device_Response_Time",
    "hour",
]

X = df_train[feature_cols]
y = df_train["Attack Type"]

imputer = SimpleImputer(strategy="median")
X_imputed = imputer.fit_transform(X)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

y_train_pred = clf.predict(X_train)
y_val_pred = clf.predict(X_val)
```

```python
from sklearn.metrics import classification_report

# Classification reports
report_train = classification_report(y_train, y_train_pred, output_dict=True)
report_val = classification_report(y_val, y_val_pred, output_dict=True)
df_train_report = pd.DataFrame(report_train).transpose()
df_val_report = pd.DataFrame(report_val).transpose()

df_train_report, df_val_report
```

```python
df_test["Timestamp"] = pd.to_datetime(df_test["Timestamp"])
df_test["hour"] = df_test["Timestamp"].dt.hour

subtask1_answer = df_test["hour"].apply(lambda h: "AM" if h < 12 else "PM")

X_test = df_test[feature_cols]

X_test_imputed = imputer.transform(X_test)
X_test_scaled = scaler.transform(X_test_imputed)

y_test_pred = clf.predict(X_test_scaled)

sub1 = pd.DataFrame(
    {"subtaskID": 1, "datapointID": df_test["ID"], "answer": subtask1_answer}
)
sub2 = pd.DataFrame(
    {"subtaskID": 2, "datapointID": df_test["ID"], "answer": y_test_pred}
)

submission = pd.concat([sub1, sub2], ignore_index=True)

submission.to_csv("02_data/submission.csv", index=False)
```

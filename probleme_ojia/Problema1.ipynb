```python
import re

import numpy as np
import pandas as pd
```

```python
df_train = pd.read_csv("01_data/train_data.csv")
df_test = pd.read_csv("01_data/test_data.csv")
```

```python
subtask_1 = len(df_train)
subtask_2 = int(
    np.floor(
        df_train[df_train["Credit_Utilization_Ratio"] >= 25][
            "Monthly_Inhand_Salary"
        ].mean()
    )
)
subtask_3 = df_train["Month"].nunique()
subtask_4 = df_train["SSN"][df_train["SSN"].str.endswith("20", na=False)].nunique()
```

```python
def findSusValues(df: pd.DataFrame) -> dict:
    susDict = {}
    for col in df.columns:
        if df[col].dtype == "object":
            foundSus = set()
            uniqueVals = df[col].unique()
            for val in uniqueVals:
                valString = str(val)
                if re.search(r"[^A-Za-z0-9.,-/ ]+", valString):
                    foundSus.add(valString)
            if foundSus:
                susDict[col] = foundSus
    return susDict


trainSusDict = findSusValues(df_train)
for col, val in trainSusDict.items():
    print(f"column:{col}\nvalue:{val}\n")
```

```python
susList = [
    "#F%$D@*&8",
    "_______",
    "__-333333333333333333333333333__",
    "!@9#%8",
    "__10000__",
    "_",
]
for col in df_train.columns:
    df_train[col] = df_train[col].replace(susList, np.nan)
df_train.head()
```

```python
numCol = [
    "Age",
    "Annual_Income",
    "Num_of_Loan",
    "Num_of_Delayed_Payment",
    "Changed_Credit_Limit",
    "Outstanding_Debt",
    "Amount_invested_monthly",
    "Monthly_Balance",
]
for col in numCol:
    df_train[col] = df_train[col].replace(r"[^0-9.-]", "", regex=True)
    df_train[col] = pd.to_numeric(df_train[col], errors="coerce")
```

```python
df_train
```

```python
def clean_name(name: str) -> str:
    if pd.isna(name):
        return name

    name = re.sub(r'"', "", name.strip())
    name = re.sub(
        r'[a-zA-Z]+[" ]*$', lambda m: "" if len(m.group(0)) <= 2 else m.group(0), name
    )
    return name.strip()


df_train["Name"] = df_train["Name"].apply(clean_name)
```

```python
trainSusDict = findSusValues(df_train)
for col, val in trainSusDict.items():
    print(f"column:{col}\nvalue:{val}\n")
df_train.head()
```

```python
def parse_credit_history_age(val: str) -> float:
    if pd.isna(val):
        return np.nan
    match = re.match(r"(\d+)\s+Years\s+and\s+(\d+)\s+Months", val)
    if match:
        years = int(match.group(1))
        months = int(match.group(2))
        return years * 12 + months
    return np.nan


df_train["Credit_History_Age"] = df_train["Credit_History_Age"].apply(
    parse_credit_history_age
)
df_train["Credit_History_Age"]
```

```python
df_train.isna().sum()
```

```python
for col in ["Occupation", "Payment_Behaviour", "Credit_Mix", "SSN"]:
    df_train[col].fillna("Unknown", inplace=True)
for col in ["Changed_Credit_Limit", "Amount_invested_monthly", "Monthly_Balance"]:
    df_train[col].fillna(df_train[col].median(), inplace=True)
df_train.isna().sum()
```

```python
# Am realizat ca coloanele de mai jos nu sunt necesare pentru modelare
df_train.drop(columns=["ID"], inplace=True)
df_train.drop(columns=["Name"], inplace=True)
df_train.drop(columns=["SSN"], inplace=True)
df_train.drop(columns=["Customer_ID"], inplace=True)
df_train
```

```python
# Am gasit foarte usor intervalele folositndu-ma de o extensie de VSCode numita Data Wrangler
df_train_filtered = df_train[
    (df_train["Age"].between(0, 60, inclusive="both"))
    & (df_train["Annual_Income"] <= 200_000)
    & (df_train["Num_Bank_Accounts"].between(0, 10, inclusive="both"))
    & (df_train["Num_Credit_Card"].between(0, 12, inclusive="both"))
    & (df_train["Num_of_Loan"].between(0, 12, inclusive="both"))
    & (df_train["Interest_Rate"] <= 40)
    & (df_train["Num_of_Delayed_Payment"].between(0, 40, inclusive="both"))
    & (df_train["Changed_Credit_Limit"].between(0, 40, inclusive="both"))
    & (df_train["Num_Credit_Inquiries"].between(0, 20, inclusive="both"))
    & (df_train["Total_EMI_per_month"].between(0, 300, inclusive="both"))
    & (df_train["Amount_invested_monthly"] <= 1000)
    & (df_train["Monthly_Balance"] <= 1000)
]

rows_removed = len(df_train) - len(df_train_filtered)
df_train = df_train_filtered

rows_removed
```

```python
df_train.info()
```

```python
from sklearn.preprocessing import LabelEncoder

non_numeric_cols = df_train.select_dtypes(include=["object"]).columns

label_encoders = {}
for col in non_numeric_cols:
    le = LabelEncoder()
    df_train[col] = le.fit_transform(df_train[col])
    label_encoders[col] = le
```

```python
df_train
```

```python
from sklearn.model_selection import train_test_split

train, valid = train_test_split(df_train, test_size=0.2, random_state=42)

X_train = train.drop(columns=["Credit_Score"])
y_train = train["Credit_Score"]

X_valid = valid.drop(columns=["Credit_Score"])
y_valid = valid["Credit_Score"]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

model = RandomForestClassifier(random_state=42, n_estimators=100)
model.fit(X_train, y_train)

y_pred_train = model.predict(X_valid)
print("Accuracy:", accuracy_score(y_valid, y_pred_train))
print(classification_report(y_valid, y_pred_train))
print(confusion_matrix(y_valid, y_pred_train))
```
